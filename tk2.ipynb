```python
import pandas as pd
import matplotlib.pyplot as plt

# Mengambil data dari file .csv
data = pd.read_csv("data_tower.csv")
df = pd.DataFrame(data)

# Membuat line plot
plt.plot(df['Ukuran'], df['Harga_Sewa'], marker='o')

# Memberikan label sumbu X dan Y
plt.xlabel('Ukuran')
plt.ylabel('Harga Sewa')

# Memberikan judul
plt.title('Hubungan antara Ukuran dan Harga Sewa')

# Menampilkan hasil akhir dari visualisasi data
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Mengambil data dari file .csv
data = pd.read_csv("data_tower.csv")
df = pd.DataFrame(data)

# Menghitung rata-rata Tarif Internet berdasarkan Tipe Bangunan
mean_prices = df.groupby('Tipe_Bangunan')['Tarif_Internet'].mean()

# Membuat visualisasi Bar
plt.bar(mean_prices.index, mean_prices)

# Memberikan label sumbu X dan sumbu Y
plt.xlabel('Tipe Bangunan')
plt.ylabel('Tarif Internet')

# Memberikan judul plot
plt.title('Perbandingan rata-rata Tarif Internet berdasarkan Tipe Bangunan')

# Menampilkan hasil akhir dari visualisasi data
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Mengambil data dari file .csv
data = pd.read_csv("data_tower.csv")
df = pd.DataFrame(data)

# Mengurutkan DataFrame berdasarkan tipe bangunan
df_sorted = df.sort_values('Tipe_Bangunan')

# Membuat line plot berdasarkan Tipe Bangunan
for tipe_bangunan in df_sorted['Tipe_Bangunan'].unique():
    df_tipe = df_sorted[df_sorted['Tipe_Bangunan'] == tipe_bangunan]
    plt.plot(df_tipe['Ukuran'], df_tipe['Harga_Sewa'], marker='o', label=tipe_bangunan)

# Memberikan label sumbu X dan Y
plt.xlabel('Ukuran')
plt.ylabel('Harga Sewa')

# Memberikan Judul
plt.title('Tren Harga Sewa berdasarkan Tipe Bangunan')

# Membuat Legend
plt.legend()

# Menampilkan hasil akhir dari visualisasi data
plt.show()
```
